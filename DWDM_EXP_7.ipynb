```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data = pd.read_csv('/content/groceries - groceries.csv')
data_encode = pd.get_dummies(data)
X = data_encode.drop('Item(s)',axis=1)
X.head()
```

```python
freq_items = apriori(X,min_support=0.1,use_colnames=True)
```
