```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
data = pd.read_csv('students_placement_data.csv')
data.head()
```

```python
data = data.drop('Roll No',axis=1)
data.head()
```

```python
data = data.drop('Gender',axis=1)
data = data.drop('Section',axis=1)
data = data.drop('SSC Percentage',axis=1)
data.head()
```

```python
data_encoded = pd.get_dummies(data,columns=['registered_for_ Placement_Training'])
X = data_encoded.drop('placement status',axis=1)
Y = data_encoded['placement status']
data_encoded.head()
```

```python
X_train,X_test,Y_train,Y_test  = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
cls = DecisionTreeClassifier()
cls
```

```python
cls.fit(X_train,Y_train)
```

```python
Y_pred = cls.predict(X_test)
Y_pred
```

```python
print("accuracy_score = ",accuracy_score(Y_test,Y_pred))
```

```python
print("classifcation_report :",classification_report(Y_test,Y_pred))
```
