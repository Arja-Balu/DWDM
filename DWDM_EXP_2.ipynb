```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('collegePlace.csv')
data.head()
```

```python
data_encoded = pd.get_dummies(data, columns=['Gender', 'Stream'])
data_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data_encoded[['Age', 'Internships', 'CGPA', 'HistoryOfBacklogs']] = scaler.fit_transform(data_encoded[['Age', 'Internships', 'CGPA', 'HistoryOfBacklogs']])
```

```python
data_encoded.head()
```

```python
X = data_encoded.drop('PlacedOrNot',axis=1)
Y = data_encoded['PlacedOrNot']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```
